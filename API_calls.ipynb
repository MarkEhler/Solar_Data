<h3> Finding Best Tilt Angle </h3>

Use one of these formulas to find the best angle from the horizontal at which the panel should be tilted:

    If your latitude is below 25째, use the latitude times 0.87.
    If your latitude is between 25째 and 50째, use the latitude, times 0.76, plus 3.1 degrees.
    If your latitude is above 50째, see Other Situations below. 
    
<i> citation: https://www.solarpaneltilt.com/ <i>

```python
def tilt_angle(lat):
    '''
    accepts a latitude and returns infor regarding suggested tilt and web address for more reading  
    '''
    # might code web address into app dashboard
    # should code this fucntion into dash with box anf field for answer
    if lat < 25:
        tilt = lat * 0.87
    elif (lat > 25) and (lat < 50):
        tilt = lat * 0.76 + 3.1
    else:
        string = print('See Link')
    string = print(f' Suggested tilt of array: {tilt} [degrees]')    
    return string
```

```python
def temp_converter(temp, f_to_c=True):
    '''
    converts an integer or float to degrees celcius
    or vice verse
    '''
    if f_to_c:
        new_temp = (temp - 32) * 5/9
    else:
        new_temp = (temp * 5/9) + 32
    return new_temp
```
