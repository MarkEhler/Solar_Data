```python
def create_df(paths, cols): 
    '''
    takes a list of paths to csv files and a list of column names
    loops through and create dataframe from seperate csv file paths
    '''
    df = pd.DataFrame()
    count = 0
    listed = list()
    for i, path in enumerate(paths):
        if count % 10 == 0: 
            print(count, end = "...")
        count += 1
        with open(path, newline='') as f:
            reader = csv.reader(f, delimiter=',')
            for l in reader:
                item = pd.DataFrame([l])
                listed.append(item)
    df = pd.concat([i for i in listed], axis=0)
    print(df.shape)
    df.columns = cols
    df.head(1)
    return df
```

<h4> Condensing Columns </h4>
We have so many columns and many of them track the same thing.  The following function is designed to handle those redundancies and save them to a final column we can add back to the dataframe

```python
global_irradiance =  ['Global Horizontal [W/m^2]', 'Global Horiz [W/m^2]', 'Global PSP [W/m^2]']

battery_DC_voltage = ['CR1000 Battery [VDC]', 'CR510 Battery [VDC]', 'Logger Battery [VDC]', 'RSR Battery [VDC]', 'CR10 Battery [VDC]']

tempurature = ['Air Temperature [deg C]',
 'Roof Temp [deg C]', 
 'CR510 Temp [deg C]', 
 'Dry Bulb Temp [deg C]']

time_zones = ['PST',
              'MST',
              'EST',
              'HST']

direct_solar = ['Direct Normal [W/m^2]',
 'Direct Normal (calc) [W/m^2]',    
 'Direct NIP [W/m^2]',]
```

```python
def condense_columns(a_list, df):
    '''
    takes a list of columns names and their parent dataframe
    loops through the list to fill null objects
    returns a dataframe where the rightmost column has the fewest nulls
    '''
    condense = pd.DataFrame(df[a_list[0]])
    
    print(f'{condense.isna().sum()}\n...')
    for idx, i in enumerate(a_list):
        if idx+1 >= len(a_list):
            break
        elif idx == 0:
            condense[a_list[0] + str(0)] = condense[a_list[0]].fillna(df[a_list[idx+1]])
        else:
            condense[a_list[0] + str(idx)] = condense[a_list[0] + str(idx-1)].fillna(df[a_list[idx+1]])
            print(f'{condense.isna().sum()}\n...')
#         test_zones = test_zones.reset_index(drop = True)
    condense = condense.iloc[:,-1]
    return condense
```

```python
minutes = condense_columns(time_zones, joined_dropped_df)
# batt = condense_columns(battery_DC_voltage, joined_dropped_df)
irradiance = condense_columns(global_irradiance, joined_dropped_df)
temp = condense_columns(tempurature, joined_dropped_df)
direct = condense_columns(direct_solar, joined_dropped_df)
```

```python
# join updated columns to existing dataframe

joined_dropped_df['Temp [deg C]'] = temp
joined_dropped_df['Minutes'] = minutes
# joined_dropped_df['Battery [VDC]'] = batt
joined_dropped_df['Direct Irradiance [W/m^2]'] = direct
joined_dropped_df['Global Irradiance [W/m^2]'] = irradiance
```

```python
def add_coordinates(coordinates, df):
    '''
    takes a dict of location keys and latitude and longitude as listed pairs as values
    takes a dataframe with columns 'location' 'Latitude [deg]' 'Longitude [deg]'
    adds hardcoded coordinates to the named locations
    outputs the updated dataframe
    '''
    for location in coordinates:
        df['Latitude [deg]'] = np.where(df['location'] == location, coordinates[location][0],
                                           df['Latitude [deg]'])
        df['Longitude [deg]'] = np.where(df['location'] == location, coordinates[location][1],
                                           df['Longitude [deg]'])
#     we won't be needing this anymore
    df.drop(['location'], axis = 1, inplace = True)
    return df
```

<h2> Changing Data Types </h2>

```python
# convert type to numeric data
def type_set(df):
    '''
    takes a dataframe
    checks for date columns and converts them to integers
    converts all other columns to floats
    returns the converted dataframe
    '''
    for col in df.columns:
        df = df.astype({col : 'float'})
    df['Year'] = df['Year'].astype('str')
    df['DOY'] = df['DOY'].astype('str')
    return df
```
